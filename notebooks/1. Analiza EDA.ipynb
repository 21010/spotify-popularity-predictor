### 3. Analiza Zmiennej Celu
Zmienną objaśnianą (target) w projekcie jest `popularity` – liczba całkowita z zakresu 0-100, określająca obecny poziom popularności utworu na platformie Spotify.

W tym kroku badam rozkład empiryczny tej zmiennej. Jest to kluczowe dla wyboru odpowiedniej strategii modelowania, ponieważ:
1. Modele regresji liniowej najlepiej radzą sobie z rozkładami zbliżonymi do normalnego (krzywa Gaussa). Silna asymetria (skośność) może sugerować konieczność transformacji zmiennej (np. logarytmowania).
2. W danych muzycznych często występuje zjawisko nadreprezentacji utworów o bardzo niskiej lub zerowej popularności (tzw. "długi ogon"). Zrozumienie proporcji między utworami niszowymi (popularność ~0) a hitami jest niezbędne do właściwej oceny skuteczności modelu (dobór metryk RMSE vs MAE).

Poniższy wykres przedstawia histogram popularności wraz z naniesioną krzywą gęstości prawdopodobieństwa (KDE).

```python
# Inicjalizacja wizualizera
viz = SpotifyVisualizer(df)

# Rozkład popularności
viz.plot_target_distribution(target_col='popularity')
```

#### Wnioski z analizy zmiennej celu (`popularity`)

1.  Dominacja Zer (Zero-Inflation Problem):
Wykres ujawnia masywny słupek przy wartości `0`. Oznacza to, że znaczna część zbioru to utwory, które nie zdobyły żadnej mierzalnej trakcji na platformie. Wskazuje to na barierę wejścia, którą utwór musi pokonać, by algorytm w ogóle zaczął go polecać.

2. Struktura Bimodalna (Dwie populacje):
    * Rozkład nie jest jednolity. Widzimy wyraźne rozdzielenie na:
        * Grupę "Cichą" (Popularność = 0) - utwory martwe algorytmicznie.
        * Grupę "Aktywną" (Popularność ~15-80) - ta część wykresu przypomina rozkład normalny z wierzchołkiem (modą) w okolicach 40-50 punktów.
    * Istnieje widoczna luka między wartościami 0 a ~10. Sugeruje to, że jeśli utwór zdobędzie jakiekolwiek odtworzenia, zazwyczaj szybko przeskakuje do poziomu 10-20 pkt.

3.  Ogon (Right Skew):
Prawa strona wykresu (popularność 80-100) opada łagodnie. Prawdziwe globalne hity są rzadkością.

#### Wpływ na projekt
   * Standardowa Regresja Liniowa może mieć tutaj problem. Będzie próbowała poprowadzić linię "pośrodku" (przez średnią ~33), co spowoduje duży błąd zarówno dla utworów zerowych (będzie zawyżać), jak i dla hitów (będzie zaniżać).
   * Sytuacja ta faworyzuje modele oparte na drzewach decyzyjnych (Random Forest, XGBoost). Drzewa naturalnie radzą sobie z takimi barierami, tworząc warunek w stylu:<br/>
          `IF (track_has_traction) THEN (predict_value) ELSE (predict_zero)`.


### 4. Analiza Korelacji i Cech Audio

W tym kroku badam zależności statystyczne między zmiennymi numerycznymi. Moim celem jest:
1. Identyfikacja Predyktorów - Znalezienie cech, które najsilniej korelują ze zmienną celu (`popularity`).
2. Detekcja Współliniowości (Multicollinearity) - Wykrycie par cech, które niosą tę samą informację. Jest to kluczowe dla stabilności modelu, ponieważ redundantne cechy mogą sztucznie zawyżać wariancję estymatorów.

#### Metodologia
Wykorzystam Korelację Spearmana. Korelacja Pearsona bada tylko zależności liniowe. W muzyce wiele zależności jest nieliniowych (np. wzrost głośności nie przekłada się liniowo na energię w nieskończoność). Spearman (oparty na rangach) lepiej wychwytuje monotoniczne zależności.

**Hipotezy do weryfikacji na wykresie:**
* Spodziewam się bardzo silnej dodatniej korelacji między **`energy`** a **`loudness`**. Z fizycznego punktu widzenia głośniejsze utwory są zazwyczaj bardziej energetyczne. Jeśli korelacja przekroczy 0.7, rozważę usunięcie jednej z tych zmiennych, aby zredukować szum.
* Spodziewam się silnej ujemnej korelacji między **`acousticness`** a **`energy`**.

```python
# Macierz korelacji Spearmana (lepsza dla zależności nieliniowych)
viz.plot_correlation_matrix()
```

#### Wnioski z analizy macierzy korelacji

1.  Potwierdzona Współliniowość (Energy vs Loudness):
    * Zgodnie z hipotezą, obserwujemy bardzo silną korelację dodatnią (zazwyczaj > 0.75) między `energy` a `loudness`.
    * W produkcji muzycznej głośność (kompresja dynamiki) jest głównym narzędziem do budowania "energii" utworu.
    * Te dwie cechy niosą tą samą informację. W przypadku modeli wrażliwych na współliniowość (np. Regresja Logistyczna/Liniowa), należałoby usunąć `loudness`. Modele oparte na drzewach (XGBoost, Random Forest), które planuje użyć, poradzą sobie z tym lepiej, wybierając jedną z nich w węzłach decyzyjnych.

2.  Oś Akustyczna (Acousticness vs Energy):
    * Widoczna silna korelacja ujemna. Utwory akustyczne rzadko są energetyczne/głośne. To naturalne rozdzielenie gatunkowe (np. Muzyka Klasyczna vs Metal).

3.  Brak "Srebrnej Kuli" dla Popularności:
    * Żadna pojedyncza cecha audio nie wykazuje bardzo silnej korelacji (powyżej 0.5) ze zmienną `popularity`. Wartości korelacji z celem są prawdopodobnie niskie (w zakresie -0.2 do 0.2).
    * "Przepis na hit" nie jest prosty. Nie wystarczy zrobić utworu "szybszego" lub "głośniejszego", by stał się popularny.
    * Zależność jest nieliniowa i wielowymiarowa. Prosta regresja liniowa oparta na jednej z cech dałaby słabe wyniki ($R^2$ bliskie 0). Konieczne jest użycie modeli, które potrafią łapać złożone interakcje między cechami (np. "Szybki ORAZ Głośny ORAZ Taneczny" = Hit), czyli **Random Forest** lub **XGBoost**.

4.  Valence (Nastrój):
    * Ciekawą obserwacją jest zazwyczaj umiarkowana korelacja `valence` z `danceability`. Utwory do tańca są zazwyczaj "weselsze" (pozytywny nastrój), choć w nowszej muzyce Pop ta zależność słabnie (popularność "smutnego dance'u").

### 5. Analiza Liniowości (Features vs Target)
W tej sekcji weryfikuje fundamentalne założenie algorytmów regresji liniowej: liniową zależność między zmiennymi niezależnymi a zmienną celu.

Generuje wykresy punktowe (Scatter Plots) z naniesioną linią regresji dla kluczowych cech audio (`danceability`, `energy`, `acousticness`, `instrumentalness`).

**Szukam:**
1. Wyraźnych trendów - Czy istnieje prosta reguła, np. "im utwór bardziej taneczny, tym bardziej popularny"? (Linia wznosząca).
2. Chmury punktów (Noise) - Jeśli punkty są rozproszone chaotycznie i linia trendu jest płaska, oznacza to, że dana cecha w izolacji jest słabym predyktorem liniowym.
3. Zależności nieliniowych - Być może "przepis na hit" znajduje się w środku skali (np. tempo umiarkowane jest lepsze niż skrajnie wolne lub szybkie).

**Oczekiwania:**
W danych socjologicznych/artystycznych rzadko spotyka się czyste zależności liniowe ($R^2 > 0.5$ dla pojedynczej cechy). Spodziewam się dużego rozrzutu (szumu), co będzie silnym argumentem za odrzuceniem prostej regresji liniowej na rzecz algorytmów nieliniowych, potrafiących modelować złożone interakcje (Random Forest / Gradient Boosting).

```python
# Wykresy punktowe dla najważniejszych cech
features_to_check = ['danceability', 'energy', 'acousticness', 'instrumentalness']
viz.plot_features_vs_target(features=features_to_check, sample_size=10000, target_col='popularity')
```

#### Wnioski z analizy

1.  Brak Silnej Zależności Liniowej:
    * Dla większości cech (`energy`, `valence`, `tempo`) chmury punktów są mocno rozproszone. Linia regresji jest często płaska lub ma minimalne nachylenie.
    * Pojedyncza cecha techniczna (np. "szybkość utworu") nie determinuje sukcesu. Nie ma prostej reguły: "zwiększ tempo o 10 BPM, a popularność wzrośnie o 5%".

2.  Instrumentalness (Kluczowa obserwacja):
    * Wykres dla `instrumentalness` wykazuje specyficzny charakter. Widoczne jest duże skupisko utworów o wartości bliskiej `0` (utwory z wokalem) rozsianych po całej skali popularności, oraz skupisko przy wartości `1` (czysta muzyka instrumentalna), które zazwyczaj ma niską popularność.
    * Jest to silny sygnał nieliniowy/binarny. Obecność wokalu jest niemal "warunkiem koniecznym" (choć niewystarczającym) do osiągnięcia statusu globalnego hitu.

3.  Danceability (Lekki Trend):
    * Jedynie w przypadku `danceability` można dostrzec delikatny trend wzrostowy. Utwory bardziej taneczne mają statystycznie nieco większą szansę na wyższą popularność, co jest zgodne z charakterystyką współczesnego rynku Pop/Hip-Hop.

#### Wpływ na projekt
* Obserwowany "szum" i brak silnych korelacji liniowych ostatecznie dyskwalifikują prostą Regresję Liniową jako główny model (spodziewamy się bardzo niskiego $R^2$).
* Złożoność relacji (np. wysoka popularność możliwa zarówno przy niskiej, jak i wysokiej energii, ale zależna od gatunku) wymusza zastosowanie modeli potrafiących mapować nieliniowe granice decyzyjne – Random Forest lub XGBoost.

Chcę sprawdzić, czy poziom popularności jest różny dla różnych gatunków. Innymi słowy: czy istnieją gatunki, w których osiągnięcie wysokiej popularności jest statystycznie łatwiejsze niż w innych.

### 8. Podsumowanie Analizy EDA i Strategia Modelowania

Przeprowadzona analiza eksploracyjna pozwoliła na zrozumienie natury danych i sformułowanie kluczowych założeń dla etapu modelowania.

#### Kluczowe Wnioski

1.  Charakterystyka Zmiennej Celu (`popularity`):
    * Rozkład jest bimodalny z silną nadreprezentacją wartości `0` (Zero-Inflation).
    * Istnieje wyraźna bariera wejścia – utwory albo nie istnieją w algorytmie (0 pkt), albo wpadają w rozkład zbliżony do normalnego (15-80 pkt).
    * Nie jest wymagana transformacja logarytmiczna celu, ale konieczne jest wybór modelu odporny na "zera".

2.  Cechy Audio i Współliniowość:
    * Wykryto silną współliniowość (**> 0.75**) między `energy` a `loudness`.
    * Pojedyncze cechy audio (tempo, danceability) wykazują słabą korelację liniową z celem, co dyskwalifikuje prostą regresję liniową jako główny model.
    * W modelach drzewiastych (XGBoost) można zostawić obie cechy (model sam wybierze lepszą w węźle). W modelach liniowych musielibyśmy usunąć jedną z nich.

3.  Kluczowe Predyktory:
    * Gatunek (`track_genre`) - Najsilniejszy predyktor. Określa "bazowy poziom" popularności. Wymaga One-Hot Encodingu.
    * Czas trwania (`duration_ms`): Zależność nieliniowa typu "dzwon". Istnieje optymalne okno (2:30 - 4:00 min).
    * Explicit Content: Treści wulgarne korelują z wyższą popularnością (przez dominację gatunków Hip-Hop/Trap).

#### Strategia Modelowania

Na podstawie powyższych wniosków, odrzucam Regresję Liniową jako docelowe rozwiązanie (spodziewam się niesatysfakcjonujących wyników ze względu na nieliniowość danych i bimodalny rozkład).

Jako możliwe modele rozważam:
1.  XGBoost (Extreme Gradient Boosting) ze względu na zdolność obsługi danych rzadkich (One-Hot Encoding gatunków), nieliniowości oraz odporność na outliery.
2.  Random Forest jako silną podstawę dla modeli nieliniowych.
